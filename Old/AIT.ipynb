Import lines from a .txt file.

```python
with open('/content/drive/MyDrive/FHNW/AI Technologies/Project_data.txt') as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
file.close()
```

Remove any excess information leaving only the data to be analysed.

```python
content = []
excess = []
for line in lines:
    if ('@' in line) or ('%' in line):
        excess.append(line)
    else:
        content.append(line)
content.pop(0)
```

Create a numpy array of the data.

```python
import numpy as np
holding_array = []
for i in range(len(content)):
    lineSplit = content[i].split(',')
    holding_array.append(lineSplit)
data = np.array(holding_array)
```

Replace strings with numerical equivalents.
Make the data array dtype int.

```python
data = np.where(data == 'yes','1', data)
data = np.where(data == 'no','0', data)
data = np.where(data == 'cholesterol','1', data)
data = np.where(data == 'diabetes','2', data)
data = np.where(data == 'heart','3', data)

data = np.where(data == 'low','1', data)
data = np.where(data == 'medium','2', data)
data = np.where(data == 'high','3', data)

data = data.astype(int)
```

Separate the initial conditions from the target outcome.

```python
init = []
target = []
for i in range(len(data)):
    tuple=[]
    for j in range(len(data[i])):
        if j==len(data[i])-1:
            target.append(data[i][j])
        else:
            tuple.append(data[i][j])
    init.append(tuple)    
```

Check the shape of init.

```python
X = np.array(init)
X.shape
```

Check the shape of target.

```python
Y = np.array(target)
Y.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, Ytrain)
Y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Ytest, Y_model)
```
